> I found this gist today, because I wanted to download kaggle dataset into google colab, but I ended up using a different approach, I hope it helps:`!mkdir /root/.kaggle !echo '{"username":"USERNAME","key":"API_KEY"}' > /root/.kaggle/kaggle.json `This way you don't need to change kaggle json file directory in the kaggle settings and now you can use kaggle to do whatever you need!`!kaggle datasets download -d owner/dataset-slug`


`캐글 홈페이지 > 사용자 프로필 > Account`를 클릭하면 Account페이지가 나옵니다. 조금 아래의 API 메뉴에 있는 `Create New API Token` 버튼을 클릭하면 `kaggle.json`파일을 다운로드 받습니다.

이 파일을 열어보면 

> "username":"USERNAME","key":"API_KEY"

가 나옵니다. 예를 들면


````
!mkdir -p /root/.kaggle
!echo '{"username":"aimldl","key":"56d3330b17d24860e916199484f8b838"}' > /root/.kaggle/kaggle.json
````

````
!chmod 600 /root/.kaggle/kaggle.json
````

파일 권한을 변경하지 않으면 에러 메세지가 발생합니다.
